### 請寫一程式可判斷udn.jl 內容中與"蔡英文"有關的新聞有多少則

```
import json

def count_udn(filename, keyword):
    count = 0

    with open(filename, 'r', encoding='utf-8') as file:

        for line in file:
            data = json.loads(line)

            if keyword in data.get('content', '').lower():
                count += 1

    return count

filename = 'udn.jl'
keyword = '蔡英文'

result = count_udn(filename, keyword)

print(f'與 "{keyword}" 有關的新聞有{result} 則')
```
